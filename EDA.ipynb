```python
import cv2
import matplotlib.pyplot as plt

# Cargar la imagen
imagen = cv2.imread('lanus.png')

# Verificar si la imagen se cargó correctamente
if imagen is None:
    print("Error al cargar la imagen 'lanus.png'. Asegúrese de que el archivo existe en el directorio actual.")
    exit()

# Convertir de BGR (formato de OpenCV) a RGB
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

# Mostrar información básica de la imagen
print(f'Dimensiones de la imagen: {imagen.shape}')
print(f'Tipo de dato: {imagen.dtype}')
```

```python
# Mostrar la imagen
plt.imshow(imagen_rgb)
plt.title('Imagen Satelital - Lanús')
plt.axis('off')
plt.show()
```

```python
# Histograma de los canales de color
canales = ('r', 'g', 'b')
for i, canal in enumerate(canales):
    hist = cv2.calcHist([imagen_rgb], [i], None, [256], [0, 256])
    plt.plot(hist, color=canal)
    plt.xlim([0, 256])
plt.title('Histograma de los Canales de Color')
plt.xlabel('Valor de Intensidad')
plt.ylabel('Cantidad de Pixeles')
plt.show()
```

```python
# Calcular estadísticas básicas
media_color = cv2.mean(imagen_rgb)
print(f'Valores medios de los canales RGB: R={media_color[0]:.2f}, G={media_color[1]:.2f}, B={media_color[2]:.2f}')
```

```python
# Mostrar histograma de intensidad en escala de grises
imagen_gris = cv2.cvtColor(imagen_rgb, cv2.COLOR_RGB2GRAY)
plt.hist(imagen_gris.ravel(), bins=256, range=[0, 256], color='gray')
plt.title('Histograma de Intensidad en Escala de Grises')
plt.xlabel('Intensidad de Pixel')
plt.ylabel('Cantidad de Pixeles')
plt.show()
```

```python
# Mostrar la imagen en escala de grises
plt.imshow(imagen_gris, cmap='gray')
plt.title('Imagen en Escala de Grises')
plt.axis('off')
plt.show()
```

```python
# Detección de bordes usando el algoritmo de Canny
bordes = cv2.Canny(imagen_gris, 100, 200)
plt.imshow(bordes, cmap='gray')
plt.title('Detección de Bordes - Canny')
plt.axis('off')
plt.show()
```
