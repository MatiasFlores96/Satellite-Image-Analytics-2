```python
import cv2
import matplotlib.pyplot as plt

# Cargar la imagen en escala de grises
image = cv2.imread('lanus.png', cv2.IMREAD_GRAYSCALE)

# Aplicar el detector de bordes de Canny con diferentes umbrales
edges_1 = cv2.Canny(image, 50, 150)
edges_2 = cv2.Canny(image, 100, 200)
edges_3 = cv2.Canny(image, 150, 250)

# Mostrar la imagen original y las imágenes con los bordes detectados
plt.figure(figsize=(10, 7))

# Imagen original
plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Imagen Original')
plt.axis('off')

# Canny con umbrales 50 y 150
plt.subplot(2, 2, 2)
plt.imshow(edges_1, cmap='gray')
plt.title('Canny (50, 150)')
plt.axis('off')

# Canny con umbrales 100 y 200
plt.subplot(2, 2, 3)
plt.imshow(edges_2, cmap='gray')
plt.title('Canny (100, 200)')
plt.axis('off')

# Canny con umbrales 150 y 250
plt.subplot(2, 2, 4)
plt.imshow(edges_3, cmap='gray')
plt.title('Canny (150, 250)')
plt.axis('off')

# Mostrar todas las imágenes
plt.tight_layout()
plt.show()
```
